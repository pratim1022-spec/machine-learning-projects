```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

import warnings
warnings.filterwarnings("ignore")
```

```python
# OUTCOME IS THE LABEL AND WE NEED TO SEPEARTE DATA FROM LABEL SEPARETLY

x = diabetes_dataset.drop(columns = 'Outcome',axis =1)  # axis = 1 for dropping a column and axis = 0 for dropping a row
y = diabetes_dataset['Outcome']
```

```python
x.head()
```

```python
y
```

```python
# STANDARDIZE THE DATA , TO BRING THE DATA WITHIN A FIXED RANGE

scaler = StandardScaler()
```

```python
standarized_data = scaler.fit_transform(x)
```

```python
print(standarized_data)
```

```python
x = standarized_data
y = diabetes_dataset['Outcome']
```

```python
print(x)
print(y)
```

```python
# TRAIN AND TEST SPLIT

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=2) # stratify = y ensures that both the training and testing sets maintain approximately the same proportion of each class as the original dataset
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
# TRAINING THE SUPPORT VECTOR MACHINE CLASSIFIER

classifier.fit(x_train,y_train)
```

```python
# MODEL EVALUATION

# ACCURACY SCORE ON TRAINING DATA

x_train_pred = classifier.predict(x_train)
training_data_accuracy = accuracy_score(x_train_pred,y_train)
```

```python
print("Accuracy score of predicting a training data : ", training_data_accuracy)
```

```python
# ACCURACY SCORE ON TEST DATA

x_test_pred = classifier.predict(x_test)
test_data_accuracy = accuracy_score(x_test_pred,y_test)
```

```python
print("Accuracy score of test data : ", test_data_accuracy)
```

```python
# MAKING A PREDICTIVE SYSTEM TO CHECK IF A PERSON HAS DIABETES OR NOT

input_data = [ (10,168,74,0,0,38,0.537,34),(13,145,82,19,110,22.2,0.245,57)]

# CONVERTING INPUT DATA INTO NUMPY ARRAYS AS IT IS MORE EFFICIENT

input_data_numpy_array = np.asarray(input_data)

# STANDARIZE THE INPUT DATA

std_data = scaler.transform(input_data_numpy_array)
#print(std_data)

# Create mapping and DataFrame directly
label_map = {0: "Not diabetic", 1: "Diabetic"}

df = pd.DataFrame(input_data, columns=[
    "Pregnancies", "Glucose", "BloodPressure", "SkinThickness",
    "Insulin", "BMI", "DiabetesPedigreeFunction", "Age"
])

# Add predictions and labels
df["Prediction"] = classifier.predict(std_data)
df["Result"] = df["Prediction"].map(label_map)

# Show results
print(df)
print(df[["Prediction", "Result"]])

for i in range(len(df)):
    print(f"Person {i+1}: {df.loc[i, 'Result']}")
```
